## Table 1: Top Tracks with ___ Vibes

 1. User selects a couple words describing the mood, style, and/or genre of music they want.
 2. Dashboard displays top X most popular songs and and top X most popular artist that show up in playlists titled with those words.

```python
playlist_words = list(df1['Unnamed: 0'])
```

```python
track_ids = list(df1.columns[1:])
```

```python
def rank_track_ids(words):
    filter_words = []
    for word in words:
        if word in playlist_words:
            filter_words.append(word)
    if len(filter_words) > 0:
        # focus on rows matching the user input word
        filter = df1['Unnamed: 0'].isin(filter_words)
        

        # sum the values in each column
        totals = df1.loc[filter, track_ids].sum()
        totals = list(totals)

        sorted_pairs = sorted(zip(totals, track_ids), reverse=True)
        totals_sorted, track_ids_sorted = zip(*sorted_pairs)
        track_ids_sorted = list(track_ids_sorted)
        totals_sorted = list(totals_sorted)
        return track_ids_sorted, totals_sorted
    else:
        return LookupError('No input words are in list of possible playlist words.')
```

```python
def get_track_info(track_ids):
    results = []
    for track_id in track_ids:
        filter = df3['track_id'] == track_id
        track_name = df3[filter]['track_name'].iloc[0]
        artists = df3[filter]['artists'].iloc[0]
        album_name = df3[filter]['album_name'].iloc[0]
        results.append((track_name,artists,album_name))
    return results
```

```python
top_x = 20

# given user's selected words
user_words = ['guilty']
print(f'Top {top_x} songs with vibes: {", ".join(user_words)}')
print('====================================================')
# rank track_ids based on selected words
track_ids, totals = rank_track_ids(user_words)

# show title, artists, album for top_x songs
top_track_ids = track_ids[:top_x]
top_tracks = get_track_info(top_track_ids)

for i in range(top_x):
    print(i+1, top_tracks[i], totals[i])
```

## Plot 2: Artist Vibes 

 3. User can also select an artist from the above list and then a word cloud appears showing frequency of words used for playlists containing their songs

```python
user_artist = 'Carly Rae Jepsen'
filter = df2['artist'] == user_artist
# Select the row (Kansas) and drop artist column
row = df2.loc[filter, df2.columns[2:]].iloc[0]

# Get top 3 words by value
top_pairs = row.sort_values(ascending=False).head(top_x)

# Plot with seaborn
plt.figure(figsize=(6,4))
sns.barplot(x=top_pairs.values, y=top_pairs.index)

plt.title(f"Top {top_x} Words for {user_artist}")
plt.xlabel("ScoreWord")
plt.ylabel("Word")
plt.show()
```

```python
from wordcloud import WordCloud
```

```python
# Extract frequencies (exclude the 'artist' column)
frequencies = df2.loc[filter, df2.columns[2:]].iloc[0].to_dict()

# Create word cloud
# wc = WordCloud(width=800, height=400, background_color="white", colormap="viridis")
# Define a color function that always returns blue
def blue_color_func(*args, **kwargs):
    return "rgb(31, 119, 180)"  # pure blue, you can adjust RGB

# Create word cloud with uniform color
wc = WordCloud(
    width=800,
    height=400,
    background_color="white",
    color_func=blue_color_func
).generate_from_frequencies(frequencies)

# Plot
plt.figure(figsize=(10, 5))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
