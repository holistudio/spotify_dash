```python
# df_t = sub_df.set_index('track_id').T.reset_index().rename(columns={'index': 'playlist_word'})

df_t = sub_df.set_index('track_id').T
df_t.columns.name = "playlist_word"
df_t.reset_index()
df_t.head()
```
